```python
income=[50000, 60000, 200000, 62000, 40000]
employment_status=["employed", "unemployed","employed", "unemployed"]
qualified_count=0
for i in range(len(income)):
  if income[i]>=50000 and employment_status[i]=="employed":
    qualified_count+=1
    print(f"person {i+1} is qualified for loan")
  else:
    print(f"person {i+1} is not qualified")
print("Number of qualified persons:", qualified_count)

```

```python

speeds = [20, 69, 80, 70 , 30, 26, 65, 30]

total_vehicles = len(speeds)
violations = 0

print("Overspeeding Vehicles (speed > 80 km/h):")

for speed in speeds:
    if speed > 80:
        print(speed, "km/h")
        violations += 1


percentage = (violations / total_vehicles) * 100

print("\nTotal Vehicles:", total_vehicles)
print("Total Violations:", violations)
print("Violation Percentage:", percentage, "%")
```
